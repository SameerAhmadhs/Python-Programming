# String Indexing 

### Access the characters in a String --> Visualizing the String Index
#### String index numbering

* Index numbers can be +ve/-ve
* +ve Index numbers always start with 0. 
* -ve Index numbers always end at -1.
* First char will have the Index num : 0 or -(len(str))
* Last char will have the Index num : -1 or len(str)-1
* Indexing is done by using the operator - [ ]
  
  ```
     Say, "a" is a string variable

   - a[index num] - extracts the single character

   - a[start:stop]- items start through stop-1

   - a[start:]- items start through the rest of the array

   - a[:stop]- items from the beginning through stop-1

   - a[:]- a copy of the whole array

   - a[::-1]- reverse the string (here -1 is the step)
   

 

````python
myname="SAMEER AHMAD"
# myname[index num]
print(myname[0])
print(myname[1])
print(myname[2])
print(myname[3])
print(myname[4])
print(myname[5])
print(len(myname))

````

````python
# Extract the first character
print('myname[0]:',myname[0])# First method
print('myname[-(len(myname))]:',myname[-len(myname)])# second method
````

````python
# Extract the last character
print("myname[11]:",myname[11])#first method
print("myname[len(myname)-1]:", myname[len(myname)-1])#second method
````

````python
# a[start:stop] - ":" - colon is the sequence operator
print('myname[0:11]',myname[0:12])
print('myname[0:]',myname[0:])
print('myname[1:6]',myname[1:6])
print('myname[3:10]',myname[3:10])
print('myname[0:1]',myname[0:1])
````

````python
# a[start:]
print('myname[0:]',myname[0:])
````

````python
# a[:stop]
print(myname[:11])
print(myname[:len(myname)])
````

````python
# a[:]
print(myname[:])
````

````python
# a[start:end:step] - where step is the increment of position; by default step = 1
print("myname[1:10:1]",myname[1:10:1])
print("myname[1:10:2]",myname[1:10:3])
print("myname[0:11:2]",myname[0:11:2])
````

````python
# Reversing the string
print("myname[::]",myname[::])
print("myname[::-1]",myname[::-1])
print("myname[::-2]",myname[::-2])
print("myname[::-3]",myname[::-3])
print("myname[::-4]",myname[::-4])
````

````python
# Using bith +ve & -ve indexing numbers
print('myname[1:-6]',myname[1:-6])
print('myname[1:-2]',myname[1:-2])
print('myname[2:-6]',myname[2:-6])
print('myname[1:-6]',myname[1:-6])
````

````python
# Using bith +ve & -ve indexing numbers
print('myname[1:-6]:', myname[1:-6])
print('myname[-6:1]:', myname[-6:1]) # reverse direction doesn't work
print('myname[1:-6:-1]:', myname[1:-6:-1])
print('myname[-6:-1]:', myname[-6:-1])
print('myname[-6:-1][::-1]:', myname[-6:-1][::-1])
````

#  String formatting in Python 

1. Using %d, %f, %s
2. Using str.format()
3. Using f-strings

````python
# Using %d, %f, %s
a,b=80,"1stDivistion"
print("If you score %d you will get %s"%(a,b))
print("If you score %f you will get %s"%(a,b))

# Using str.format()
print("If you score {}% you will get {} class".format(a,b))

# f-strings (Formatted strings)
print(f"If you score {a}% you will get {b} class")
````
