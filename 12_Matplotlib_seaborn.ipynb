```python
print(plt.style.available)
```

```python
# mobile sales by year

# generate 4 arrays with 12 random values each 
np.random.seed(111)
sales2017 = np.round(np.sort(np.random.uniform(10,45,size=12)),2)
sales2018 = np.round(np.sort(np.random.uniform(24,55,size=12)),2)
sales2019 = np.round(np.sort(np.random.uniform(18,65,size=12)),2)
sales2020 = np.round(np.sort(np.random.uniform(15,75,size=12)),2)

month_no = np.arange(1,13)

plt.figure(figsize=(15,5),dpi=100)

plt.plot(month_no,sales2017, label='2017 sales', color='#464196',ls = '--', lw = 5, marker = '.', markersize = '30')
plt.plot(month_no, sales2018, label = '2018 Sales', color = '#a0025c', ls = '-', lw = 3, marker = 'o', markersize = '10')
plt.plot(month_no, sales2019, label = '2019 Sales', color = 'b', ls = '-.', lw = 4, marker = 'D', markersize = '10')
plt.plot(month_no, sales2020, label = '2020 Sales', color = 'teal', ls = ':', lw = 2, marker = 'H', markersize = '10')

# adding information to the chart
plt.title('Mobiles sales from 2017-2020', fontsize = 17, color = 'red')
plt.xticks(month_no)
plt.xlabel('month number')
plt.yticks(np.arange(5,100,15))
plt.ylabel('Sales in lakhs')

plt.legend(loc = 'best')
plt.show();


```

## 4. Scatter plot

```python
np.random.seed(123)

temp = np.sort(np.random.uniform(21,50, size =14).round(2))
AC_sales = np.sort(np.random.randint(1000, 10000, size=14))
TV_sales = np.random.randint(5000, 10000, size=14)
```

```python
plt.scatter(temp,AC_sales,marker='D',s = 50,c = 'teal')
plt.xlabel('Temp', fontsize=16)
plt.ylabel('AC Sales', fontsize=16)
plt.title('Scatter plot - Temp vs AC Sales',fontsize=20)
plt.show();
```

```python
plt.scatter(temp, TV_sales, marker = 'H', s = 50, c = 'red')
plt.xlabel('Temp', fontsize=16)
plt.ylabel('TV Sales', fontsize=16)
plt.title('Scatter plot - Temp vs TV Sales',fontsize=20)
plt.show();
```

## Scatter plot by categories

```python
plt.scatter(temp, AC_sales, marker = 'D', s = 50, c = 'teal', label = 'AC')
plt.scatter(temp, TV_sales, marker = 'H', s = 50, c = 'red', label = 'TV')


plt.xlabel('Temp', fontsize=16)
plt.ylabel('Sales', fontsize=16)
plt.title('Scatter plot - Temp vs AC & TV Sales',fontsize=15)
plt.legend(loc = 'lower right')
plt.show();
```

## 6. Box plot 

```python
# box plot
temp = np.sort(np.random.uniform(21,50,size = 30).round(2))
ac_sale = np.sort(np.random.randint(1000,10000,size=30).round(2))
TV_sales = np.random.randint(5000, 10000, size=30)
```

```python
c = 'red'
plt.boxplot(ac_sale, patch_artist = True,
           boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color='aqua'));
```

```python
c = 'blue'
plt.boxplot([ac_sale, TV_sales], patch_artist = True, vert = 0,labels = ['AC_sales', 'TV_sales'],
           boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color='aqua')); # horizontal box plot
```

## 4. Box plot

```python
plt.figure(figsize=(15,5))
sns.boxplot(data = diamond_df, x = 'price');
```

```python
diamond_df.price.describe()
```

```python
diamond_df.price.quantile(q = [0.25,0.5,0.75])
```

```python
Q1, Q2, Q3 = tuple(diamond_df.price.quantile(q = [0.25,0.5,0.75]).values)
```

```python
print(Q1,Q2,Q3)

IQR = Q3-Q1
print(IQR)

# UB = Q3 + 1.5*IQR
UB = Q3 + (1.5*IQR)
# LB = Q1-1.5*IQR
LB = Q1 - (1.5*IQR)

print(UB, LB)
```

```python
# using the UB & LB values we need to find the edges of the dataset

# the vertical line above Q3 is - the max value in the dataset less than UB
diamond_df[diamond_df.price < UB]['price'].max()
# the vertical line below Q1 is - the min value in the dataset greater than LB
diamond_df[diamond_df.price > LB]['price'].min()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data = diamond_df, x = 'price', color = 'yellowgreen')

plt.axvline(Q1,lw = 1, ls = '--')
plt.axvline(Q2,lw = 1, ls = '--')
plt.axvline(Q3,lw = 1, ls = '--')

plt.axvline(11883,lw = 1, ls = '--')
plt.axvline(326,lw = 1, ls = '--');
```
