```python
# for 1000 rows
name = []
price = []
change_1h = []
change_24h = []
change_7d = []
volume_24h = []
marketcap = []
Page_no = []

for i in range(1,50):
    
    #Time taken by page to extract the data(starting point)
    start_time = time.time()
    
    #Website
    URL = 'https://www.coingecko.com/?page='+str(i)
    
    #Requesting to get data from website
    response = requests.get(URL)
    
    soup = BeautifulSoup(response.content,'html.parser')
    
    results = soup.find('table',{'class':'table-scrollable'}).find('tbody').find_all('tr')
    
    for result in results:
        #name
        try:
            name.append(result.find('a',{'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
        except:
            name.append('/na')

        #price
        try:
            price.append(result.find('td',{'class':'td-price price text-right pl-0'}).get_text().strip())
        except:
            price.append('/na')

        #1h change
        try:
            change_1h.append(result.find('td',{'class':'td-change1h change1h stat-percent text-right col-market'}).get_text().strip())
        except:
            change_1h.append('/na')

        #24h change
        try:
            change_24h.append(result.find('td',{'class':'td-change24h change24h stat-percent text-right col-market'}).get_text().strip())
        except:
            change_24h.append('/na')

        #7d change
        try:
            change_7d.append(result.find('td',{'class':'td-change7d change7d stat-percent text-right col-market'}).get_text().strip())
        except:
            change_7d.append('/na')

         #24h volume
        try:
            volume_24h.append(result.find('td',{'class':'td-liquidity_score lit text-right col-market'}).get_text().strip())
        except:
            volume_24h.append('/na')

        #marketcap
        try:
            marketcap.append(result.find('td',{'class':'td-market_cap cap col-market cap-price text-right'}).get_text().strip())
        except:
            marketcap.append('/na')
            
        Page_no.append(i)
    
    #End time
    end_time = time.time()
    print('Page {} completed in {} seconds'.format(i, end_time-start_time))    
    
```

```python
crypto_df = pd.DataFrame({'Coin_Name':name,'Price':price,'1h':change_1h,'24h':change_24h,'7d':change_7d,'24h_Volume':volume_24h,'Marketcap':marketcap,'Page_no':Page_no})
```
